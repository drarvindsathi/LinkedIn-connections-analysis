```python
import pandas as pd
import os
```

```python
df_name_gender_combined = pd.read_csv('Name Gender Probability.csv')
```

```python
df_linkedin_names = pd.read_csv(os.path.join('Basic_LinkedInDataExport_04-19-2021', 'Connections_cleaned.csv'))
df_linkedin_names_sample = df_linkedin_names[0:20]
df_linkedin_names_first_names = df_linkedin_names_sample[['First Name']].applymap(lambda x: str(x).split()[0])
df_linkedin_names_first_names
```

```python
df_linkedin_names_first_names_sorted = df_linkedin_names_first_names[['First Name']].sort_values(by=['First Name'])
df_linkedin_names_first_names_sorted
```

```python
df_linkedin_names_combined = df_linkedin_names_first_names_sorted.pivot_table(index=['First Name'], aggfunc= 'size').reset_index()
df_linkedin_names_combined = df_linkedin_names_combined.rename(columns={'First Name': 'Name', 0: 'Count'})
df_linkedin_names_combined['Name'] = df_linkedin_names_combined['Name'].str.lower()
df_linkedin_names_combined
```

```python
df_linkedin_names_combined = df_linkedin_names_combined.merge(df_name_gender_combined, how='left', on='Name',
          suffixes=('_left', '_right'))
df_linkedin_names_combined
```
